$$

$$

```python
# Análise gráfica da interesção entre o bandido e o xerife

fig, ax = plt.subplots()
plt.title('Perseguição ao ladrão de banco')
plt.xlabel('Tempo (em minutos)')
plt.ylabel('Distância (em km)')
ax.set_xlim([0, 40])
ax.set_ylim([0, 100])
ax.plot(t, d_b, c='green')
ax.plot(t, d_x, c='brown')
plt.axvline(x=30, color='purple', linestyle='--')
_ = plt.axhline(y=75, color='purple', linestyle='--')
```

# Modelo de Neurônio Artificial 
O Neurônio é a unidade fundamental de processamento em sistemas Redes Neurais Artificiais. Desenvolvido em 1943, por McCulloch-Pitts, o modelo artificial do neurônio é representa o processo de ativação de um neurônio devido a um dado estímulo, seja ele dos neurônios receptores dos sentidos (olfato, paladar, tato, visão audição, etc) ou vindo de outros neurônios. 

Matematicamente, este estímulo é representado por um vetor de entradas **x** no qual
$$
\mathbf{x} = 
\begin{bmatrix}
x_1 \\
x_2 \\
\vdots \\
x_m
\end{bmatrix}
$$
O modelo do neurônio é dividido em camada de entrada, somatório, função de ativação e saída, mostrado na Figura 1.

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider
#import ipywidgets as widgets

# Função que plota a reta com os parâmetros escolhidos
def plot_reta(w1, w2, b):
    x = np.linspace(-10, 10, 400)
    y = - x * w1/w2 + b/w2

    plt.figure(figsize=(8, 5))
    plt.plot(x, y, label=f'y = x({w1:.2f}/{w2:.2f}) + {b:.2f}/{w2:.2f}', color='blue')
    plt.axhline(0, color='black', lw=0.5)
    plt.axvline(0, color='black', lw=0.5)
    plt.grid(True)
    plt.ylim(-20, 20)
    plt.xlim(-10, 10)
    plt.title('Reta Interativa: y = a(w_1/w_2) + b/w_2')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.show()

# Sliders interativos
interact(plot_reta,
         w1=FloatSlider(min=-10, max=10, step=0.1, value=1.0, description='Coef. w_1'),
         w2=FloatSlider(min=.0001, max=10, step=0.1, value=1.0, description='Coef. w_2'),
         b=FloatSlider(min=-30, max=30, step=0.1, value=0.0, description='Coef. b'))
```
